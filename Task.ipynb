```python
stoplist=os.listdir() 
stoplist
```

```python
all_stop_word=[]
```

```python
for i in stoplist:
    with open(i,'r') as text:
        aa=text.readlines()
        all_stop_word+=aa
        print(all_stop_word)
        
        
```

```python
stop_words=[i.strip() for i in  all_stop_word]
```
